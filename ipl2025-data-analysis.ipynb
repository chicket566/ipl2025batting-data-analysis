```python
import numpy as np
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("IPL2025Batters.csv")
```

```python
print(df)
```

```python
print(df.columns)
```

```python
filtered=df[df["Runs"]>=300]
```

```python
sns.barplot(data=filtered,y="Player Name",x="Runs",hue="Team")
```

```python
group=df.groupby("Team")
```

```python
series=group["Runs"].sum()
print(series)
```

```python
sns.barplot(y=series.index,x=series.values,)
plt.xlabel("RUNS")
```

```python
series=series.sort_values(ascending=False)
print(series)
```

```python
sns.barplot(y=series.index,x=series.values)
plt.xlabel("RUNS")
```

```python
boundaries=df.groupby("Team")
total_boundaries=boundaries["4s"].sum()+boundaries["6s"].sum()
total_boundaries=total_boundaries.sort_values(ascending=False)
print(total_boundaries)
```

```python
sns.barplot(x=total_boundaries.values,y=total_boundaries.index)
plt.xlabel("total boundaries")
```
